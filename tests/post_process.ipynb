````python
with open("../data/processed/fix_header/chapter7.md", "r", encoding="utf-8") as f:
    text = f.read()

text[:1000]
````

````python
lines = text.split("\n")
lines[:10]
````

````python
# Get REAL HEADERS (Ignore comments in code blocks)
headers = []
ignore = False

for line in lines:
    if line.startswith("#") and not ignore:
        headers.append(line)
    
    # If we encounter a code block (```), we stop appending headers by switching the ignore bool to True
    # When we see the closing ``` sign of the code block, the ignore bool is switched to False again
    # And we continue appending real headers
    if line.strip() == "```":
        ignore = not ignore

headers
````

````python
# Add a special symbol at the beginning of each header to differentiate from code comments
SPECIAL_CHAR = "|"

header_idx = 0
cur_header = headers[0]
num_headers = len(headers)

for i, line in enumerate(lines):
    if line == cur_header:
        lines[i] = SPECIAL_CHAR + cur_header
        header_idx += 1
        if header_idx == num_headers:
            break
        cur_header = headers[header_idx]

lines[:10]
````

After adding `|` at the beginning of each header, we can later define headers for `MarkdownHeaderTextSplitter` as follows
```python
HEADERS = [
    ("|# ", "Header 1"),
    ("|## ", "Header 2"),
]
```
and so on. You can choose your own SPECIAL_CHAR, as long as it doesn't cause further confusion

````python
# Update the header list as well

headers = ["|" + header for header in headers]
headers
````

````python
# Get the list of headers that might need merging
# We'll only merge deepest sections
def get_depth(header):
    depth = 0
    for c in header[1:]:    # Ignore the beginning special char
        if c == "#":
            depth += 1
        else:
            return depth
        
# Get headers of the maximum depth
max_depth = max(get_depth(header) for header in headers)
max_depth
````

````python
# Build sibling relationships (not necessarily based on header numbers)
prev_depth = 0
prev_header = None

siblings = {}

for header in headers:
    cur_depth = get_depth(header)
    # If 2 consecutive headers are both of the deepest level, they are siblings
    if cur_depth == max_depth and cur_depth == prev_depth:
        # Add those 2 headers to the siblings dictionary
        if prev_header not in siblings:
            siblings[prev_header] = {}
        siblings[prev_header]["subsequent"] = header
        if header not in siblings:
            siblings[header] = {}
            siblings[header]["previous"] = prev_header
    prev_depth = cur_depth
    prev_header = header

siblings
````

````python
# Build a dictionary to check if a header has been merged
merged = {header: False for header in headers}
merged
````

````python
from langchain_text_splitters import MarkdownHeaderTextSplitter


headers_to_split_on = [
    ("|#", "Header 1"),
    ("|##", "Header 2"),
    ("|###", "Header 3"),
    ("|####", "Header 4"),
]

# Header splits
markdown_splitter = MarkdownHeaderTextSplitter(
    headers_to_split_on,
    strip_headers = False,
    )

md_header_splits = markdown_splitter.split_text("\n".join(lines))
md_header_splits[:3]
````

````python
def normalize_header(header):
    return header.lstrip("|").lstrip("#").strip()
    
def canonical(h):
    return h.strip()
clean_siblings = {}

for k, v in siblings.items():
    clean_k = normalize_header(k)
    clean_siblings[clean_k] = {}

    if "previous" in v:
        clean_siblings[clean_k]["previous"] = normalize_header(v["previous"])

    if "subsequent" in v:
        clean_siblings[clean_k]["subsequent"] = normalize_header(v["subsequent"])

clean_siblings
````

````python
def get_deepest_header(doc):
    headers = list(doc.metadata.values())
    return headers[-1].strip() if headers else None

header_to_idx = {}
for idx, doc in enumerate(md_header_splits):
    deepest = get_deepest_header(doc)
    if deepest:
        header_to_idx[normalize_header(deepest)] = idx
````

````python
def merge_metadata(meta1, meta2):
    merged = {}

    for k in set(meta1) | set(meta2):
        v1, v2 = meta1.get(k), meta2.get(k)

        if v1 == v2:
            merged[k] = v1
        else:
            merged[k] = [v for v in [v1, v2] if v is not None]

    return merged
````

````python
MERGE_LENGTH = 1024

merged_doc = []
used = set()

for i, doc in enumerate(md_header_splits):
    if i in used:
        continue
    
    current_content = doc.page_content
    current_metadata = doc.metadata.copy()
    current_header = normalize_header(get_deepest_header(doc))
    
    if len(current_content) >= MERGE_LENGTH:
        merged_doc.append(doc)
        used.add(i)
        continue
    
    siblings = clean_siblings.get(current_header, {})
    prev = siblings.get("previous")
    sub = siblings.get("subsequent")

    target_idx = None

    for j, other_doc in enumerate(md_header_splits):
        if j != i and j not in used:
            other_header = normalize_header(get_deepest_header(other_doc))
            
            if other_header == current_header:
                target_idx = j
                break

    if target_idx is None and prev in header_to_idx:
        idx = header_to_idx[prev]
        if idx not in used and idx < i:
            target_idx = idx

    if target_idx is None and sub in header_to_idx:
        idx = header_to_idx[sub]
        if idx not in used and idx > i:
            target_idx = idx
    
    if target_idx is not None:
        target_doc = md_header_splits[target_idx]
        target_content = target_doc.page_content

        if target_idx < i:
            merged_content = target_content + "\n\n" + current_content
        else:
            merged_content = current_content + "\n\n" + target_content

        current_metadata = merge_metadata(current_metadata, target_doc.metadata)
        used.add(target_idx)
    else:
        merged_content = current_content

    # cập nhật doc
    doc.page_content = merged_content
    doc.metadata = current_metadata

    merged_doc.append(doc)
    used.add(i)
````

````python
import re

def finalize_merged_documents(docs):
    """
    Hàm xử lý hậu kỳ cho danh sách các document đã merge.
    - Sắp xếp metadata Header.
    - Làm sạch ký tự đặc biệt trong nội dung.
    - Đưa metadata vào nội dung trang.
    """
    processed_docs = []
    
    for doc in docs:
        # lấy và sort header
        header_keys = sorted(
            [k for k in doc.metadata.keys() if re.match(r'Header\s*\d+', k)],
            key=lambda x: int(re.search(r'\d+', x).group())
        )
        
        # đưa header về str
        def flatten_to_str(v):
            if isinstance(v, list):
                
                flat = []
                def _rec(l):
                    for item in l:
                        if isinstance(item, list): _rec(item)
                        else: flat.append(str(item).strip())
                _rec(v)
                return " & ".join(list(dict.fromkeys(flat)))
            return str(v).strip()

        # update metadata header
        clean_metadata = {}
        header_values = []
        for k in header_keys:
            val_str = flatten_to_str(doc.metadata[k])
            clean_metadata[k] = val_str
            header_values.append(val_str)
            
        # xóa | và #
        clean_text = re.sub(r'^[|#\s]+', '', doc.page_content, flags=re.MULTILINE).strip()
        
        # tạo chuỗi header
        hierarchy_str = " > ".join(header_values)
        
        # chèn header vào page content
        if hierarchy_str:
            final_content = f"NGỮ CẢNH: {hierarchy_str}\n{'-'*30}\n{clean_text}"
        else:
            final_content = clean_text
            
    
        doc.page_content = final_content
        doc.metadata = clean_metadata
        processed_docs.append(doc)
        
    return processed_docs

final_results = finalize_merged_documents(merged_doc)
````
